```python
#Pipeline Function
#it connects model with its necessary preprocessing steps,allowing us to directly input any text and get an intelligible answer
```

```python
from transformers import pipeline
```

```python
classifier=pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
classifier=pipeline("sentiment-analysis")
classifier(["I don't like exams.","I hate this subject.","I love reading novels."])
```

```python
classifier=pipeline("zero-shot-classification")
classifier(
    "Ronaldo",
    candidate_labels=["education", "politics", "business","Football"],
)
```

```python
generator=pipeline("text-generation")
generator("Programming is the")
```

```python
generator=pipeline("text-generation",model="distilgpt2")
generator(
    "Programming is the",
    max_length=30,
    num_return_sequences=2,
    )
```

```python
#Mask Filling
unmasker=pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.",top_k=2)
```

```python
#Token Classification
ner=pipeline("ner",grouped_entities=True)
ner("My name is Dipti and I work at hugging face in Brooklyn")
```
